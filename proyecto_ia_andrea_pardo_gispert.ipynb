### Descripción del Conjunto de Datos

El conjunto de datos contiene 12 columnas en total, divididas en 6 columnas numéricas y 6 columnas categóricas. A continuación, describimos cada una de las columnas y sus características:

```python
# Vamos a poner los valores de HDI en función de "country" y "year"
human_dev_index_df = human_dev_index_df.pivot(index='country', columns='year', values='HDI for year')
human_dev_index_df.head()
```

```python
human_dev_index_df.info()
```

vamos a añadir la información que falta en la columna "HDI for year" usando la información que hemos conseguido en el dataframe "human_dev_index"
Aunque también hay valores nulos en este dataframe, la intención es que reduzca el numero de huecos

```python
# Asegurarse de que el año sea string en human_dev_index_df
human_dev_index_df.columns = human_dev_index_df.columns.astype(str)

# Iterar sobre cada fila del dataset y añadir el valor de HDI que falte
# Excluir las filas con año anterior a 1985, que es el primer año comun en los dos datasets
for row in suicide_rates_df.itertuples():
    if row.year >= 1985:
        try:
            human_dev_index = human_dev_index_df.loc[row.country, str(row.year)]
            suicide_rates_df.at[row.Index, 'HDI for year'] = human_dev_index
        except KeyError:
            pass

# Comprobar si ha disminuido el número de valores NA
missing_hdi = suicide_rates_df.loc[suicide_rates_df['HDI for year'].isna(), ['country', 'year']].drop_duplicates()
print(missing_hdi)
print(f"Total missing HDI values: {len(missing_hdi)}")
```

El numero de países y años con huecos en el HDI se ha reducido, hemos pasado de 1624 huecos a 460. No tenemos más datos para rellenar estos huecos.
podemos deducir de nuestros datos y de la bibliografia consultada (programa de naciones unidas) que el HDI va aumentando con el paso de los años. asi que, podriamos usar el HDI mínimo de cada país para rellenar huecos anteriores.

```python
# Agrupamos el HDI por país y filtramos el valor mínimo
min_human_dev_index_df = suicide_rates_df[['country', 'HDI for year']].groupby('country').agg({"HDI for year": "min"})
min_human_dev_index_df.head()
```

```python
# Vamos a convertir el dataframe en un diccionario
min_human_dev_index_dict = min_human_dev_index_df.to_dict()

# Mapeamos los valores de HDI que nos faltan con el diccionario
mask = suicide_rates_df['HDI for year'].isna()
suicide_rates_df.loc[mask, 'HDI for year'] = suicide_rates_df.loc[mask, 'country'].apply(lambda x: min_human_dev_index_dict['HDI for year'][x])
```

Hemos reducido mucho los valores nulos en HDI for year. Para los valores que nos faltan vamos a usar el IDH promedio de cada pais por año

```python
mean_hdi = human_dev_index_df.mean()
# Verificar los años en mean_hdi
print("Años en mean_hdi:", mean_hdi.index)

# Asegurarse que los años son cadenas
mean_hdi.index = mean_hdi.index.astype(str) 
```

```python
# Crear mask bool para huecos en HDI (después de 1985)
mask_aft_1985 = (suicide_rates_df['HDI for year'].isna()) & (suicide_rates_df['year'] >= 1985)

# Verificar los años que están causando problemas
missing_years = suicide_rates_df.loc[mask_aft_1985, 'year'].apply(lambda x: str(x)).unique()
print("Años faltantes en mean_hdi:", [year for year in missing_years if year not in mean_hdi.index])
```

```python
# Obtener la media del año más cercano disponible (en este caso, 1990)
mean_hdi_1990 = mean_hdi['1990']

# Imputar los años faltantes con la media del año 1990
def impute_missing_hdi(year):
    year_str = str(year)
    if year_str in mean_hdi.index:
        return mean_hdi[year_str]
    else:
        return mean_hdi_1990

suicide_rates_df.loc[mask_aft_1985, 'HDI for year'] = suicide_rates_df.loc[mask_aft_1985, 'year'].apply(impute_missing_hdi)

# Mostrar las primeras filas del DataFrame para inspeccionar los resultados
suicide_rates_df.head()

```

### Vamos a preparar los datos para la fase de entrenamiento

preparar los datos correctamente es crucial para asegurar que el modelo de machine learning funcione de manera óptima. Se van a seguir los pasos generales para preparar un conjunto de datos antes del entrenamiento:

1. **Creación de la columna Destino**

    Vamos a definir la variable de salida o "target" que el modelo va a intentar predecir

    - Identificamos la columna del conjunto de datos que representaria la variable objetivo. En nuestro caso no existe y vamos a crearla en este paso usando la información de suicidios/100k pop.
   
    - Nuestra variable objetivo va a ser el alto/bajo riesgo de suicidio "suicide_risk".
   
    - Si el valor de suicidios/100k pop es mayor que la media de suicidios/100k pop, clasificaremos el riesgo de suicidio como alto, y si se da el caso contrario, lo clasificaremos como bajo.
   
2. **Creación de una división de Entrenamiento/Validación /Prueba**

   vamos a dividir el conjunto de datos en subconjuntos para asegurar que el modelo no se sobreajuste y pueda generalizar bien.

   Dividimos el conjunto en tres pasos:
   
   - Entrenamiento (Training Set): Usado para entrenar el modelo.
     
   - Validación (Validation Set): Usado para ajustar los hiperparámetros y prevenir el sobreajuste.
     
   - Prueba (Test Set): Usado para evaluar el rendimiento final del modelo.

### Escalamiento de Características Numéricas

Para garantizar que ninguna característica numérica tenga un impacto desproporcionado en la pérdida del modelo, y para facilitar el proceso de entrenamiento, es fundamental escalar las características numéricas a un rango uniforme. En este caso, escalaremos las características numéricas al rango de valores entre 0 y 1 utilizando la técnica de **normalización**.

#### Diferencia entre Normalización y Estandarización

- **Normalización (Min-Max Scaling)**: Es el proceso de reescalar los valores de las características para que caigan dentro de un rango específico, típicamente entre 0 y 1. Esto se realiza utilizando la siguiente fórmula:
  \[
  X_{\text{normalizado}} = \frac{X - \text{min}(X)}{\text{max}(X) - \text{min}(X)}
  \]
  La normalización es útil cuando no se asume ninguna distribución específica de los datos y se quiere mantener la interpretabilidad de los valores dentro de un rango determinado.

- **Estandarización (Z-score Normalization)**: Es el proceso de reescalar los datos para que tengan una media de 0 y una desviación estándar de 1. Esto se logra con la fórmula:
  \[
  X_{\text{estandarizado}} = \frac{X - \text{media}(X)}{\text{desviación estándar}(X)}
  \]
  La estandarización es adecuada cuando se asume que los datos siguen una distribución normal y es necesario trabajar con propiedades estadísticas como la media y la desviación estándar.

Para este proyecto, se optó por la **normalización** a un rango de [0, 1] para las columnas numón y el seguimiento del proceso de preprocesamiento de datos.

```python
# Crear una instancia del scaler
scaler = MinMaxScaler()

# Ajustar el scaler a los datos numéricos del DataFrame completo
scaler.fit(suicide_rates_df[numeric_columns])

# Transformar los datos en el conjunto de entrenamiento
train_inputs[numeric_columns] = scaler.transform(train_inputs[numeric_columns])

# Transformar los datos en el conjunto de validación
val_inputs[numeric_columns] = scaler.transform(val_inputs[numeric_columns])

# Transformar los datos en el conjunto de prueba
test_inputs[numeric_columns] = scaler.transform(test_inputs[numeric_columns])

# Mostrar una descripción estadística de los datos escalados en el conjunto de entrenamiento

print(train_inputs[numeric_columns].describe())
```

### Codificación de Columnas Categóricas

Para entrenar modelos de machine learning, es necesario transformar los valores de las columnas categóricas en números, ya que la mayoría de los algoritmos no pueden trabajar directamente con datos categóricos. Existen varias técnicas para realizar esta transformación, como la **codificación de etiquetas** (Label Encoding) y la **codificación one-hot** (One-Hot Encoding).

En este proyecto, utilizaremos **OneHotEncoder** de `sklearn.preprocessing`. Este método convierte cada valor categórico en una nueva columna, asignando un valor de 1 o 0 (True/False) a la columna correspondiente. La ventaja de esta técnica es que evita asignar un peso indebido a un valor categórico sobre otro, algo que puede ocurrir con la codificación de etiquetas.

#### Proceso de Codificación One-Hot

1. **Verificación de Valores Únicos**: Comenzamos revisando el número de valores únicos en cada columna categórica para tener una idea del número de columnas que se crearán.


2. **Instanciar OneHotEncoder**: Creamos una instancia de `OneHotEncoder`, configurada para manejar categorías desconocidas durante la transformación usando `handle_unknown='ignore'`.


3. **Generación de Nombres de Columnas Codificadas**: Utilizando el método `get_feature_names`, generamos los nombres de las columnas para cada categoría individual, y los agregamos a `train_inputs`, `val_inputs` y `test_inputs`.


4. **Eliminación de Columnas Categóricas Textuales**: Finalmente, eliminamos las columnas categóricas originales, de modo que solo queden datos numéricos.

Este enfoque asegura que el modelo no interprete incorrectamente las relaciones entre los datos categóricos, proporcionando una representación adecuada de estas características en el conjunto de datos.


```python
# Verificar el número de valores únicos en cada columna categórica
unique_values = suicide_rates_df[categorical_columns].nunique()
print(unique_values)
```

```python
# Instanciar OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore').fit(suicide_rates_df[categorical_columns])
```

```python
# Generar nombres de columnas codificadas
encoded_columns = list(encoder.get_feature_names_out(categorical_columns))

# Codificar las columnas categóricas en los conjuntos de entrenamiento, validación y prueba
train_inputs[encoded_columns] = encoder.transform(train_inputs[categorical_columns]).toarray()
val_inputs[encoded_columns] = encoder.transform(val_inputs[categorical_columns]).toarray()
test_inputs[encoded_columns] = encoder.transform(test_inputs[categorical_columns]).toarray()
```
